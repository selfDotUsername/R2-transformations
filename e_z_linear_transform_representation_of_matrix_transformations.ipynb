```
import numpy as np
from matplotlib import pyplot as plt

```

```
y = np.linspace(-3.15,3.15,40)
x = np.linspace(-3.15,3.15,40)

def grid(a,b):
  ab = []
  for i in range(len(a)):
    ab.append([])
    for j in range(len(b)):
      ab[i].append([a[i],b[j]])
  return(ab)

xy = grid(x,y)



#print(xy[100][100])
```

```
def transform(matrix, L):
  result = []
  for i in range(len(matrix)):
    result.append([])
    for j in range(len(matrix[i])):
      u = matrix[i][j][0]
      v = matrix[i][j][1]
      result[i].append([  u + L * ( np.e**u * np.cos(v) - u),  v + L * ( np.e**u * np.sin(v)  - v) ])
  return(result)

#xy3 = transform(xy)


  
```

```
frames = 10
fig, ax = plt.subplots(frames+1,figsize=(10,50))

def drawgrid(a,plot):
  for i in range(len(a)-1):
    for j in range(len(a[i])-1):
      ax[plot].plot([a[i][j][0],a[i][j+1][0]],[a[i][j][1],a[i][j+1][1]])
      ax[plot].plot([a[i][j][0],a[i+1][j][0]],[a[i][j][1],a[i+1][j][1]])







L = 1 / (frames)
for i in range(frames):
  drawgrid(transform(xy, L-1/frames), i)
  L += 1 / frames

drawgrid(transform(transform(xy,1),1),10)

#drawgrid(xy1,0)
#drawgrid(xy2,1)
#drawgrid(xy3, 0)
#plt.show
```
